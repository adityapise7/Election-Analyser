```python
import pandas as pd
education_ord = [
    'Not Gone to School', 
    'Upto 9th', 
    '10th Pass', 
    '12th Pass', 
    'Graduate', 
    'Post-Graduate', 
    'Professional Education'
]
age_ord = [
    '18-19', 
    '20-29', 
    '30-39', 
    '40-49', 
    '50-59', 
    '60-69', 
    '70 & Above'
]
transformer = ColumnTransformer(transformers=[
    ('ord', OrdinalEncoder(categories=[education_ord, age_ord]), ['Education', 'Age_Group']),
    ('nom', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), 
     ['Gender', 'Geography', 'Occupation', 'Caste'])
], remainder='passthrough')
```

```python
from sklearn.model_selection import train_test_split
```

```python

X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Voted_Party']),df['Voted_Party'],test_size=0.25)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
lr_pipe = Pipeline([
    ('preprocessing', transformer),
    ('scaler', StandardScaler(with_mean=False)),
    ('model', LogisticRegression(
        solver='lbfgs',
        max_iter=1000,
        random_state=42
    ))
])
lr_pipe.fit(X_train, y_train)
y_pred_lr = lr_pipe.predict(X_test)
print(f" Accuracy: {accuracy_score(y_test, y_pred_lr):.2f}")
```
